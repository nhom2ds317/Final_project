```python
df = pd.read_csv("/content/drive/MyDrive/DS317/daxuly.csv")
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df
```

```python
# Lọc dữ liệu hợp lệ (không phải NaN, 0, 'Unknown', và 'socc_tienganh' != 0)
valid_tienganh = df['tienganh'].notna() & (df['tienganh'] != '0') & (df['tienganh'] != 'Unknown') | (df['socc_tienganh'] != 0)

# Nhóm theo 'khoahoc' và đếm số dòng hợp lệ
valid_counts = df[valid_tienganh].groupby('khoahoc').size()

# Nhóm theo 'khoahoc' và tính tổng số dòng mỗi khóa học (tổng số sinh viên)
total_counts = df.groupby('khoahoc').size()

# Nhóm theo 'khoahoc' và đếm số lượng sinh viên có chứng chỉ (sinh viên có 'socc_tienganh' != 0)
students_with_certificate = df[df['socc_tienganh'] != 0].groupby('khoahoc').size()

# Tính tỷ lệ phần trăm dữ liệu hợp lệ cho mỗi khóa học
valid_percentage = (valid_counts / total_counts) * 100

# Tính tỷ lệ phần trăm thiếu dữ liệu cho mỗi khóa học
missing_percentage = ((total_counts - valid_counts) / total_counts) * 100

# Tạo dataframe kết quả với tất cả các thuộc tính yêu cầu
result = pd.DataFrame({
    'Total Students': total_counts,
    'Students with Certificate': students_with_certificate,
    'Valid Data %': valid_percentage,
    'Missing Data %': missing_percentage
})

# In kết quả
result
```
